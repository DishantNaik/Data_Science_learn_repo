```python
# Filtering data
# Traditional way
bools = []
for val in movies.duration:
    if val >= 200:
        bools.append(True)
    else:
        bools.append(False)
print(bools[:5])
```

```python
is_long = pd.Series(bools)
```

```python
is_long.head()
```

```python
movies[is_long]
```

```python
# Filtering can be done without using FOR loop as well
is_long_new = movies['duration'] >= 200
is_long_new.head()
```

```python
movies[is_long_new]
```

```python
# A more shorter way
movies[movies['duration'] >= 200]
```

```python
# To select perticular col/cols after filter
movies.loc[movies['duration'] >= 200,['genre','title']]
```

```python
# Applying multiple filters
# NOTE: Only symbols of logical oprratos can be used like '&' for 'AND' and '|' for 'OR'
movies.loc[(movies['duration'] >= 200) | (movies['genre'] == 'Crime')]
```

```python
# Multiple ORs
movies[movies.genre.isin(['Crime','Drama','Action'])]
```
